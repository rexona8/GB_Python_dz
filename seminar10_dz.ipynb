```python
import random
```

```python
lst += ['human'] * 10
```

```python
random.shuffle(lst)
```

```python
import pandas as pd
```

```python
data = pd.DataFrame({'whoAmI':lst})
```

```python
pd.get_dummies(data['whoAmI']).head(10)
```
